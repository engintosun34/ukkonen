```python
import pandas as pd
```

```python
df = pd.read_csv("rows.csv")
```

```python
df.head()
```

```python
dfc = df[['Product','Issue','Company','State','Complaint ID','ZIP code']].copy()
```

```python
dfc.dropna(inplace=True)
```

```python
dfc.reset_index(drop=True, inplace=True)
```

```python
dfc.head(30)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
column_list = ['Product', 'Issue', 'Company']
for k in column_list:
  dfc[k] = dfc[k].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

```python
colon_listesi = ["Product","Issue","Company"]
n_isaret=[",",".","/","'"]
for n in n_isaret:
  for i in colon_listesi:
    for k in range(len(dfc['Product'])):
      if dfc[i][k].find(n) != -1:
        dfc[i][k] = dfc[i][k].replace(n,"")
      
```
